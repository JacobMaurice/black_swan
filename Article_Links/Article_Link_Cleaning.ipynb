```python
import pandas as pd

# Import links to articles
articles_2006_2007 = pd.read_csv("Articles_2006-2007.csv")
articles_2008_01_06 = pd.read_csv("Articles_2008_01-06.csv")
articles_2008_07_10 = pd.read_csv("Articles_2008_07-10.csv")
articles_2008_11_12 = pd.read_csv("Articles_2008_11-12.csv")

all_articles = pd.concat([articles_2006_2007, articles_2008_01_06, articles_2008_07_10, articles_2008_11_12], axis=0)

print(all_articles.describe())

# Remove duplicates based on title, keeping the first occurrence
unique_articles = all_articles.drop_duplicates(subset=['Title'])

# Check to see if it worked
print(unique_articles.describe())

# Save as csv
unique_articles.to_csv("NYT_Article_Links.csv", index=False)

# Save only the Headlines
unique_headlines = unique_articles['Title']
unique_headlines.to_csv("Pre-Crash_Headlines.csv", index=False)
```
